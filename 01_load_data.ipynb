```python
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus
import os

# This handles special characters like @ and * in password
password = quote_plus("Devadeep*277353@1")
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/fraud_detection')

file_csv = 'C:/FraudProject/data/creditcard.csv'
file_xlsx = 'C:/FraudProject/data/creditcard.xlsx'

if os.path.exists(file_csv):
    df = pd.read_csv(file_csv)
    print("Reading CSV file...")
elif os.path.exists(file_xlsx):
    df = pd.read_excel(file_xlsx)
    print("Reading Excel file...")
else:
    print("❌ File not found! Check the path.")

df.columns = [c.lower() for c in df.columns]
print(f"✅ Loaded shape: {df.shape}")
print(df.head())
```

```python
print(f"Loading {len(df)} rows into PostgreSQL... please wait (1-3 mins)")
df.to_sql('transactions', engine, if_exists='append', index=False, chunksize=1000)
print("✅ Data loaded successfully!")
```
